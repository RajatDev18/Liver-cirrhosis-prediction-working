```python
num_cols = ['N_Days','Age','Bilirubin','Cholesterol','Albumin','Copper','Alk_Phos','SGOT','Tryglicerides','Platelets','Prothrombin']
cat_cols = ['Status','Drug','Sex','Ascites','Hepatomegaly','Spiders','Edema','Stage']
```

# Univariate Analysis

```python
for col in num_cols:
    plt.figure(figsize=(10,6))
    sns.histplot(df[col],kde=True,color='skyblue')
    plt.title(f"Distribution of {col}")
    plt.show()
    sns.boxplot(x=df[col])
    plt.title(f"Box Plot of {col}")
    plt.show()
```

```python
for col in cat_cols:
    plt.figure(figsize=(10,6))
    sns.countplot(x=df[col])
    plt.title(f"Count plot of {col}")
    plt.show()
```

# Bivariate Analysis

Numerical vs Numerical

```python
sns.scatterplot(x="Bilirubin", y="Albumin", data=df)
plt.show()
```

Numerical vs Target

```python
sns.boxplot(x="Stage", y="Bilirubin", data=df)
plt.show()
```

Categorical vs Target

```python
pd.crosstab(df["Sex"], df["Stage"], normalize="index")
```

```python
sns.countplot(x="Stage", hue="Sex", data=df)
plt.show()
```

Correlation Matrix

```python
corr = df.corr(numeric_only=True)
corr['Stage']
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(corr, annot=False, cmap="coolwarm")
plt.show()
```

# Multivariate Analysis

```python
sns.pairplot(df, vars=["Bilirubin", "Albumin", "Platelets"], hue="Stage")
plt.show()
```

```python
sns.boxplot(x="Stage", y="Albumin", hue="Sex", data=df)
plt.show()
```
